```python
# dylan's work
from scipy.io import arff
from IPython.display import display
import os, sys
import pandas as pd
import seaborn as sns
import numpy as np 
```

```python
def arff2df (paths):    
    data_array = []    
    for i in range(len(paths)):
        data_array.append((paths[i] ,pd.DataFrame(arff.loadarff(paths[i])[0])))   
        
    return data_array
        
def get_filenames (path):
    dirs = os.listdir(path)
    dirs.sort()
    
    dirs_array = []
    for i in range(len(dirs)):
        dirs_array.append(path + dirs[i])
        
    return dirs_array
```

```python
# importing data to pandas dataframe
path = "data/"
paths = get_filenames(path)
df_years = arff2df(paths) #array of tuples (pathname, dataframe) representing year1 to year5

for i in range(len(df_years)):
    # changing y to 0 & 1 
    df_years[i][1]["class"] = df_years[i][1]["class"].astype("category").cat.codes
    # adding another column for year
    df_years[i][1]["year"] = i + 1
    # printing values 
    print("Path: {0}, Shape: {1}, Bankrupt: {2}, Not Bunkrupt: {3}".format(df_years[i][0], df_years[i][1].shape
                                                            ,((df_years[i][1]["class"] == 1).sum())
                                                            ,((df_years[i][1]["class"] == 0).sum())))
    
# append dataframes 
df_bankruptcy = pd.concat([i[1] for i in df_years], axis=0)

# print to check concat corresponds to the imported data
for i in range(len(df_years)):
    print("Year:{0} Total:{1}".format(i + 1 , (df_bankruptcy["year"] == i + 1).sum()))
print("Total instances in dataframe:" , df_bankruptcy.shape[0])

display(df_bankruptcy.head())
```
